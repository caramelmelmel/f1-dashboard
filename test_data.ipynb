```python
import fastf1

# Enable caching (recommended)
#fastf1.Cache.enable_cache('cache')

# Load session data
session = fastf1.get_session(2025, 'Brazil', 'FP1')  # year, grand prix name, session (FP1, FP2, FP3, Q, R)
session.load()

# Get lap times
laps = session.laps

print("Session loaded successfully!")
print(f"Total laps: {len(laps)}")
print(f"\nFastest lap:")
print(laps.pick_teams('Mercedes')) 
```

```python
import fastf1
import pandas as pd

now = pd.Timestamp.utcnow()
year = now.year

schedule = fastf1.get_event_schedule(year)

candidates = []
for _, event in schedule.iterrows():
    for idx in range(1, 6):
        session_name = event.get(f'Session{idx}')
        session_date = event.get(f'Session{idx}Date')
        if pd.isna(session_name) or pd.isna(session_date):
            continue
        if session_date < now:  # already started, so not live
            candidates.append((session_date, event['EventName'], session_name))

if not candidates:
    raise RuntimeError("No completed sessions yet this season.")

_, gp_name, session_label = max(candidates, key=lambda item: item[0])
session = fastf1.get_session(year, gp_name, session_label)
session.load()

results = session.results[['Position', 'DriverNumber', 'Abbreviation',
                           'TeamName', 'Points', 'Status']]
print(results)
```
